```
# Part 2.1
def hydropathy_calculator(window_size):
  aaString = ""
  for i in range(len(contentList)):
    aaString += contentList[i]
  windowList = [(aaString[i:i+window_size]) for i in range(0, len(aaString), window_size)]
  hpList = []
  for i in range(len(windowList)):
    hp = 0
    for j in range(len(windowList[i])):
      if windowList[i][j] == 'A':
        hp += 1.8  
      elif windowList[i][j] == 'I':
        hp += 4.5
      elif windowList[i][j] == 'M':
        hp += 1.9
      elif windowList[i][j] == 'T':
        hp -= 0.7
      elif windowList[i][j] == 'N':
        hp -= 3.5
      elif windowList[i][j] == 'K':
        hp -= 3.9
      elif windowList[i][j] == 'S':
        hp -= 0.8
      elif windowList[i][j] == 'R':
        hp -= 4.5
      elif windowList[i][j] == 'L':
        hp += 3.8
      elif windowList[i][j] == 'P':
        hp -= 1.6
      elif windowList[i][j] == 'H':
        hp -= 3.2
      elif windowList[i][j] == 'Q':
        hp -= 3.5
      elif windowList[i][j] == 'V':
        hp += 4.2
      elif windowList[i][j] == 'D':
        hp -= 3.5
      elif windowList[i][j] == 'E':
        hp -= 3.5
      elif windowList[i][j] == 'G':
        hp -= 0.4
      elif windowList[i][j] == 'F':
        hp += 2.8
      elif windowList[i][j] == 'Y':
        hp -= 1.3
      elif windowList[i][j] == 'C':
        hp += 2.5
      elif windowList[i][j] == 'W':
        hp -= 0.9
    totalHP = hp / window_size
    hpList.append(totalHP)
  plt.plot(hpList)
    #print("Average hydropathy of {}. window: {}".format(i+1, totalHP))
hydropathy_calculator(5)
```
